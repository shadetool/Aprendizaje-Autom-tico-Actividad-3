```python
import numpy as np
import matplotlib.pyplot as plt

# 1.- Definimos la funci칩n
def f(x):
    return 3*x + 4*np.exp(-4*x)
```

```python
# 2.- Generamos un rango de valores de x
x = np.linspace(-2, 2, 400)
```

```python
# 3.- Calculamos los valores de y
y = f(x)
```

```python
# 4.- Graficar la funci칩n
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 3x + 4e^(-4x)')
plt.title('Gr치fico de la funci칩n')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5.- Definimos la funci칩n
def f(x):
    return 5*x*(3*x + 1) - 8*x**2 - 3*x
```

```python
# 6.- Generamos un rango de valores de x
x = np.linspace(-10, 10, 400)
```

```python
# 7.- Definimos los valores de y
y = f(x)
```

```python
# 8.- Graficamos la funci칩n
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 5x(3x+1) - 8x^2 - 3x')
plt.title('Gr치fico de la funci칩n')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Importa la biblioteca SymPy
import sympy as sp

# Define las variables simb칩licas
x = sp.symbols('x')

# Define las funciones
f1 = 4*x**4 + 6*x**2 + sp.exp(4*x**2)
f2 = 1 / (2*x**3) + 8*x

# Calcula la segunda derivada de la primera funci칩n (f1)
f1_second_derivative = sp.diff(f1, x, 2)

# Calcula la segunda derivada de la segunda funci칩n (f2)
f2_second_derivative = sp.diff(f2, x, 2)

# Imprime las segundas derivadas
print("Segunda derivada de f1:")
print(f1_second_derivative)

print("\nSegunda derivada de f2:")
print(f2_second_derivative)
```
