```python
# 4.- Evaluaremos si la gráfica obtenida tiene semejanzas con la distribución normal y obtén la media, la mediana y la desviación estándar de los datos
import numpy as np

# Cálculos
media = np.mean(temp_array)
mediana = np.median(temp_array)
desviacion_estandar = np.std(temp_array)

print(f'Media: {media}')
print(f'Mediana: {mediana}')
print(f'Desviación estándar: {desviacion_estandar}')
```

```python
  # 5.- Dibujamos la gráfica de la distribución gaussiana inferida por los datos, en caso de que corresponda, y evaluamos  la suposición realizada en el punto anterior.

from scipy.stats import norm

# Rango de temperaturas
temperaturas = np.linspace(min(temp_array), max(temp_array), 100)

# Distribución gaussiana
distribucion_gaussiana = norm.pdf(temperaturas, media, desviacion_estandar)

plt.hist(temp_array, bins=20, alpha=0.75, density=True, label='Datos reales')
plt.plot(temperaturas, distribucion_gaussiana, 'r-', label='Distribución gaussiana')
plt.title('Histograma de Temperaturas con Distribución Gaussiana')
plt.xlabel('Temperatura')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```
